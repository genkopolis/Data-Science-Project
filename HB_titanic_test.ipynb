```python
def input_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass == 1:
            return 41
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
df['Age'] = df[['Age', 'Pclass']].apply(input_age, axis = 1)
df['Age']
```

```python
df.dropna(inplace = True)
```

```python
sns.heatmap(df.isnull(), yticklabels = False, cbar = False, cmap = 'viridis')
```

```python
df
```

```python
Sex = pd.get_dummies('Sex')
Sex = pd.get_dummies(df['Sex'], drop_first = True)
Sex
```

```python
Embarked = pd.get_dummies(df['Embarked'], drop_first = True)
Embarked
```

```python
df2 = pd.concat([df,Sex,Embarked], axis = 1)
df2
```

```python
df2.drop(['Name', 'Sex', 'Ticket', 'Embarked'], axis = 1, inplace = True)
df2
```

```python
df2.drop(['PassengerId'], axis = 1, inplace = True)
df2
```

```python
X = df2.drop('male', axis = 1)
X
```

```python
y = df2['male']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train,Y_train)
```

```python
Predictions = logmodel.predict(X_test)
Predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print( classification_report(Y_test,Predictions))
```

```python
print(confusion_matrix(Y_test,Predictions))
```
